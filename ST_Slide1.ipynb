```python
# Bibliotecas
import numpy as np                    # Biblioteca para cálculos numéricos e geração de dados aleatórios
import pandas as pd                   # Biblioteca para manipulação de dados em tabelas (DataFrame, Series)
import matplotlib.pyplot as plt       # Biblioteca para criação de gráficos
from statsmodels.tsa.api import seasonal_decompose   # Função para decompor séries temporais (tendência, sazonalidade, ruído)
from statsmodels.tsa.stattools import adfuller      # Teste ADF (verifica estacionariedade da série)
import statsmodels.api as sm          # Biblioteca estatística para análise de séries temporais
```

```python
# Cria uma série temporal normalmente distribuída
np.random.seed(42)                    # Define uma "semente" fixa para que os números aleatórios sejam sempre os mesmos (reprodutibilidade)
myts = np.random.normal(size=60)      # Gera 60 valores aleatórios com distribuição normal (média 0, desvio padrão 1)
```

```python
# Convertendo para série temporal com frequência mensal (12)
dates = pd.date_range(start='2012-01', end='2017-01', freq='M')   # Cria um índice de datas mensais de jan/2012 até jan/2017
myts_series = pd.Series(myts, index=dates)                        # Cria uma série temporal (pandas Series) associando valores aos meses
myts_series.plot(title='Série Temporal Aleatória')                # Plota a série temporal com título
plt.show()
```
