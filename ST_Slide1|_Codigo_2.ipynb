```python
# Importando uma série temporal de um arquivo no formato CSV
import csv                            # Biblioteca para manipulação de arquivos CSV
```

```python
def ler_csv(nome_arquivo):            # Função para ler um arquivo CSV
    try:
        with open(nome_arquivo, mode='r', encoding='utf-8') as arquivo:   # Abre o arquivo em modo leitura
            leitor_csv = csv.reader(arquivo)             # Cria o leitor do arquivo CSV
            linhas = list(leitor_csv)                    # Converte o conteúdo em lista de linhas
            return linhas                                # Retorna as linhas lidas
    except FileNotFoundError:                            # Caso o arquivo não seja encontrado
        print(f"Erro: Arquivo '{nome_arquivo}' não encontrado.")
        return None
    except Exception as e:                               # Caso aconteça qualquer outro erro
        print(f"Ocorreu um erro ao ler o arquivo: {e}")
        return None
```
