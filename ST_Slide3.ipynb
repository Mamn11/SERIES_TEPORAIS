```python
# Libraries
import numpy as np # Importa a biblioteca numpy para operações numéricas, apelidando-a de np. [cite: 18]
import pandas as pd # Importa a biblioteca pandas para manipulação e análise de dados, apelidando-a de pd. [cite: 19]
from statsmodels.datasets import sunspots # Importa o conjunto de dados sunspots da biblioteca statsmodels. [cite: 20]
import matplotlib.pyplot as plt # Importa a biblioteca matplotlib para a criação de gráficos, apelidando-a de plt. [cite: 21]
import seaborn as sns # Importa a biblioteca seaborn para visualizações estatísticas, apelidando-a de sns. [cite: 22]
```

```python
# Loading dataset Sun Spots
data = sunspots.load_pandas().data['SUNACTIVITY'] # Carrega o conjunto de dados de manchas solares e seleciona a coluna 'SUNACTIVITY'. [cite: 24]
print(data) # Exibe o conteúdo da série temporal. [cite: 25]
```

```python
# Class of object
print(type(data)) # Exibe o tipo de objeto da variável 'data'. [cite: 27]
```

```python
# Min-Max values
print("Max:", data.max()) # Calcula e exibe o valor máximo da série temporal. [cite: 29]
print("Min:", data.min()) # Calcula e exibe o valor mínimo da série temporal. [cite: 30]
```

```python
# Mean, Median, Summary and Length
print("Mean:", data.mean()) # Calcula e exibe a média da série temporal. [cite: 33]
print("Median:", data.median()) # Calcula e exibe a mediana da série temporal. [cite: 34]
print("Summary: \n", data.describe()) # Exibe um resumo estatístico completo da série. [cite: 35]
print("Length:", len(data)) # Exibe o número de observações na série temporal. [cite: 35]
```

```python
# Start and End of the series
print("Start:", data.index[0]) # Exibe o índice inicial da série, que representa o primeiro ano. [cite: 37]
print("End:", data.index[-1]) # Exibe o índice final da série, que representa o último ano. [cite: 37]
```

```python
# Cycle: Calculate the difference between the peaks
autocorr = data.autocorr(lag=12) # Calcula a autocorrelação com um lag de 12, que é a periodicidade conhecida do ciclo de manchas solares. [cite: 38]
print(f"Cycle autocorrelation at lag 12: {autocorr:.2f}") # Exibe o valor da autocorrelação. [cite: 39]
```

```python
# 1749 until december of 1763
sun2 = data.loc[1749:1763] # Cria um subconjunto da série temporal para o período de 1749 a 1763. [cite: 41]
print("\nSubset 1749-1763:") # Imprime uma linha para indicar o início da exibição do subconjunto. [cite: 41]
print(sun2) # Exibe o conteúdo do subconjunto criado. [cite: 41]
```

```python
# Load sunspots data
data = sunspots.load_pandas().data # Recarrega o conjunto de dados completo. [cite: 44]
sunspots_series = data['SUNACTIVITY'] # Atribui a coluna 'SUNACTIVITY' a uma nova variável. [cite: 45]
sunspots_series.index = data['YEAR'] # Define a coluna 'YEAR' como o índice da série temporal para plotagem. [cite: 46]
```

```python
# Time series plot
plt.figure(figsize=(12, 6)) # Cria uma figura com um tamanho especificado. [cite: 48]
sunspots_series.plot() # Plota a série temporal como um gráfico de linha. [cite: 49]
plt.title('Sunspots Time Series') # Adiciona um título ao gráfico. [cite: 50]
plt.xlabel('Year'); plt.ylabel('Sunspot Activity') # Define os rótulos dos eixos x e y. [cite: 51]
plt.show() # Exibe o gráfico. [cite: 52]
```

```python
# Histogram
plt.figure(figsize=(8, 6)) # Cria uma nova figura para o histograma. [cite: 54]
plt.hist(sunspots_series, bins=30, edgecolor='black') # Cria um histograma com 30 barras e bordas pretas. [cite: 55]
plt.title('Sunspots Distribution') # Adiciona um título ao histograma. [cite: 56]
plt.xlabel('Sunspot Activity'); plt.ylabel('Frequency') # Define os rótulos dos eixos do histograma. [cite: 57]
plt.show() # Exibe o histograma. [cite: 57]
```

```python
# Boxplot
plt.figure(figsize=(8, 6)) # Cria uma nova figura para o boxplot. [cite: 60]
sns.boxplot(y=sunspots_series) # Cria um boxplot vertical da série temporal usando a biblioteca seaborn. [cite: 61]
plt.title('Sunspots Boxplot') # Adiciona um título ao boxplot. [cite: 62]
plt.ylabel('Sunspot Activity') # Define o rótulo do eixo y. [cite: 63]
plt.show() # Exibe o boxplot. [cite: 64]
```

```python
# ggplot-like autoplot (using seaborn for prettier visualization)
plt.figure(figsize=(12, 6)) # Cria uma nova figura para o gráfico de linha do seaborn. [cite: 66]
sns.lineplot(x=sunspots_series.index, y=sunspots_series) # Cria um gráfico de linha com a biblioteca seaborn. [cite: 67]
plt.title('Sunspots Time Series') # Adiciona um título ao gráfico. [cite: 68]
plt.xlabel('Year'); plt.ylabel('Sunspot Activity') # Define os rótulos dos eixos. [cite: 69]
plt.show() # Exibe o gráfico. [cite: 69]
```

```python
# Create artificial monthly data for demonstration
dates = pd.date_range(start='1749-01-01', end='1763-12-31', freq='ME') # Cria um intervalo de datas com frequência mensal. [cite: 91]
monthly_data = pd.Series(np.random.normal(sun2.mean(), sun2.std(), size=len(dates)), index=dates) # Cria uma série temporal artificial com dados mensais. [cite: 92]
```

```python
# Aggregate to annual mean
annual_mean = monthly_data.resample('YE').mean() # Reamostra os dados mensais para obter a média anual. [cite: 94]
```

```python
# Plot
plt.figure(figsize=(12, 6)) # Cria uma figura para o gráfico dos dados agregados. [cite: 96]
annual_mean.plot() # Plota a série de média anual. [cite: 97]
plt.title('Aggregated Annual Mean (Demonstration)') # Adiciona um título. [cite: 98]
plt.xlabel('Year'); plt.ylabel('Mean Sunspot Activity') # Define os rótulos dos eixos. [cite: 99]
plt.show() # Exibe o gráfico. [cite: 99]
```
